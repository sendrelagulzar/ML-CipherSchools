# Merge & Groupby

```python
cust_df=pd.read_csv('customer_data.csv')
prod_df=pd.read_csv('product_data.csv')
pur_df=pd.read_csv('purchase_data.csv')

print(cust_df.head())
print('*'*40)
print(prod_df.head())
print('*'*40)
print(pur_df.head())
```

# Merge table

merge 2 table
inner join pur and cust df with cust_id

```python
df=pd.merge(pur_df, cust_df, how='inner', on='cust_id')
df
```

merge other table

```python
df2=pd.merge(df, prod_df, how='inner', on='prod_id')
df2.head()
```

```python
df2.prod_name.value_counts()
```

```python
df3=df2.groupby('prod_name')
df3.first()
```

```python
df3.groups.keys()
```

```python
df3.get_group('Machine Learning')
```

```python
df3.describe()
```

# Numpy Broadcasting

```python
arr=np.array([[1,2,3],[4,5,6]])
arr2=np.array([1,2,3])
arr+arr2
```

```python
arr=np.array([[1,2,3],[4,5,6]])
arr2=np.array([[1],[2]])
arr+arr2
```

```python
arr=np.array([[1,2,3],[4,5,6]])
arr2=np.array([1])
arr+arr2
```

```python
arr=np.array([1,2,3,4,5])
arr2=np.array([1,2,3,4])
arr+arr2
```

```python
arr=np.array([1,2,3,4,5])
arr2=np.array([[1],[2],[4]])
arr+arr2
```
